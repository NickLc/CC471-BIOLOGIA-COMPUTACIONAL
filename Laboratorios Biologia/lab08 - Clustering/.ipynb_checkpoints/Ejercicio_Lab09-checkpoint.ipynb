##### Análisis de Clustering de expresión génica durante crecimeinto diáuxico en levaduras.



### 1. Leer archivos 

```r
MiTabla = read.csv("data/diauxic.txt", sep="\t", row.names="NAMES")
```

### 2. Estandarizacion de datos

```r
MiTablaSTD = t(scale(t(MiTabla)))
MiTablaSTD
```

### 3. Calcular una matriz de distancias
Creamos utilizando el metodo euclidiano

```r
MisDistancias = dist(MiTabla, method="euclidean")
MisDistancias
```

### 4. Agrupacion de genes
Creamos el clustering con el metodo __Single Complete__.

```r
MiClusteringJerarquico = hclust(MisDistancias, method = "complete")
```

### 5. Graficar el dendograma

```r
plot(MiClusteringJerarquico)
```

### 6. Exportando resultados

```r
jpeg("data/resultados/dendograma_single_complete.jpeg") #alternativamente >jpeg("MiImagen.jpeg")
plot(MiClusteringJerarquico) # el plot se escribe en el PDF (no se muestra en pantalla!)
dev.off() # cierra el archivo, los próximos plots volverán a ser mostrados en pantalla
```

### Matriz de distancias usando Correlación

```r
DistanciaCorr = as.dist(1-cor(t(MiTabla)))
DistanciaCorr
```

```r
MiClusteringJerarquicoCorr = hclust(DistanciaCorr, method = "complete")
plot(MiClusteringJerarquicoCorr)
```

### Exportando resultados

```r
jpeg("data/resultados/dendograma_correlacion.jpeg") #alternativamente >jpeg("MiImagen.jpeg")
plot(MiClusteringJerarquicoCorr) # el plot se escribe en el PDF (no se muestra en pantalla!)
dev.off() # cierra el archivo, los próximos plots volverán a ser mostrados en pantalla
```

## 7. Cortes en el Arbol

__Respuesta__
Cortamos el arbol en 2 partes, usando el argumento __k=2__

En el __Cluster 1__ se encuentran las secuencias del 0 al 400 aprox y para el __Cluster 2__ se encuentran la secuencias del 400 al 800 aprox con 2 secuencias adicionales entre 210, 360 aproximadamente.

```r
MiCorte = cutree(MiClusteringJerarquico, k=2) 
plot(MiCorte)
```

## 8. Heatmap

__Respuesta__

Las secuencias del __Cluster 1__ disminuyen su expresión a medida que pasa el tiempo , mientras las secuencias del __Cluster 2__ aumentan su expresión a mediada que aumenta el tiempo.
